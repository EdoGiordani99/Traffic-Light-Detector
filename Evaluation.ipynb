```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

##Comparing Models Trainings

```python
train_histories = []
valid_histories = []

for i in range(6):

    # Importing training histories
    path = 'trainings/'+str(i+1)+'/train_history'
    try:
        f = open(path, 'rb')
        train_histories.append(pickle.load(f))
        
    except:
        train_histories.append(np.zeros(50))
        print(f"Training: file {i+1} not accessible")

    # Importing validation histories
    path = 'trainings/'+str(i+1)+'/valid_history'
    try:
        f = open(path, 'rb')
        valid_histories.append(pickle.load(f))
        
    except:
        valid_histories.append(np.zeros(50))
        print(f"Validation: file {i+1} not accessible")
```

```python
def plot_histories(histories, title, legend, range): 
    fig = plt.figure()
    ax = plt.axes()

    for hist in histories:

        # creating x axis
        x = np.linspace(0, len(hist), len(hist))

        ax.plot(x, hist);

    ax.legend(legend)
    ax.set_title(title, loc='center')
    plt.ylim(range) 
```

```python
plot_histories(histories = train_histories, 
               title = 'Training Loss', 
               legend = np.arange(len(train_histories))+1, 
               range = [0, 1])
```

```python
plot_histories(histories = valid_histories, 
               title = 'Validation Loss', 
               legend = np.arange(len(train_histories))+1, 
               range = [0.4, 0.9])
```
